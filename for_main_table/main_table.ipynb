```python
# 세종
dfsj = pd.DataFrame(pd.concat([df2015.iloc[0,:],df2016.iloc[0,:],df2017.iloc[0,:],df2018.iloc[0,:],df2019.iloc[0,:],df2020.iloc[0,:],df2021.iloc[0,:],df2022.iloc[0,:]],axis=1))
dfsj.columns=(['2015','2016','2017','2018','2019','2020','2021','2022'])
dfsj = dfsj.astype(float).astype(int)
```

```python
# 울산
dfus = pd.DataFrame(pd.concat([df2015.iloc[1,:],df2016.iloc[1,:],df2017.iloc[1,:],df2018.iloc[1,:],df2019.iloc[1,:],df2020.iloc[1,:],df2021.iloc[1,:],df2022.iloc[1,:]],axis=1))
dfus.columns=(['2015','2016','2017','2018','2019','2020','2021','2022'])
dfus = dfus.astype(float).astype(int)
```

```python
# 인천
dfic = pd.DataFrame(pd.concat([df2015.iloc[2,:],df2016.iloc[2,:],df2017.iloc[2,:],df2018.iloc[2,:],df2019.iloc[2,:],df2020.iloc[2,:],df2021.iloc[2,:],df2022.iloc[2,:]],axis=1))
dfic.columns=(['2015','2016','2017','2018','2019','2020','2021','2022'])
dfic = dfic.astype(float).astype(int)
```

```python
# 전남
dfcn = pd.DataFrame(pd.concat([df2015.iloc[3,:],df2016.iloc[3,:],df2017.iloc[3,:],df2018.iloc[3,:],df2019.iloc[3,:],df2020.iloc[3,:],df2021.iloc[3,:],df2022.iloc[3,:]],axis=1))
dfcn.columns=(['2015','2016','2017','2018','2019','2020','2021','2022'])
dfcn = dfcn.astype(float).astype(int)
```

```python
fig, ax = plt.subplots(2,2,figsize=(12,10))
for i in dfsj.index:
    x = dfsj.columns
    y = dfsj.loc[i,:]
    label = i
    ax[0,0].set_title("세종")
    ax[0,0].plot(x,y,label=label)
    ax[0,0].get_yaxis().get_major_formatter().set_scientific(False)
    ax[0,0].legend(loc='center left')
    ax[0,0].set_ylabel('판매전력량')
for i in dfus.index:
    x = dfus.columns
    y = dfus.loc[i,:]
    label = i
    ax[1,0].set_title("울산")
    ax[1,0].plot(x,y,label=label)
    ax[1,0].get_yaxis().get_major_formatter().set_scientific(False)
    ax[1,0].legend(loc='center left')
    ax[1,0].set_ylabel('판매전력량')
for i in dfic.index:
    x = dfic.columns
    y = dfic.loc[i,:]
    label = i
    ax[0,1].set_title("인천")
    ax[0,1].plot(x,y,label=label)
    ax[0,1].get_yaxis().get_major_formatter().set_scientific(False)
    ax[0,1].legend(loc='center left')
for i in dfcn.index:
    x = dfcn.columns
    y = dfcn.loc[i,:]
    label = i
    ax[1,1].set_title("전남")
    ax[1,1].plot(x,y,label=label)
    ax[1,1].get_yaxis().get_major_formatter().set_scientific(False)
    ax[1,1].legend(loc='center left')  
fig.show()
```
